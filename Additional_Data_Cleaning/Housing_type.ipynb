```python
import matplotlib
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import datetime as dt
```

```python
housing_df = pd.read_csv("Resources/DC_Properties.csv")
```

```python
housing_df.head()
```

```python
housing_df2 = housing_df.drop(["Unnamed: 0", "BLDG_NUM", "NATIONALGRID", "X", "Y"], axis=1)
housing_df2.head()
```

```python
housing_df2['YEAR'] = pd.DatetimeIndex(housing_df2['SALEDATE']).year
housing_df2['MONTH'] = pd.DatetimeIndex(housing_df2['SALEDATE']).month
housing_df2.head()
```

```python
housing_df2.columns
```

```python
housing_df3 = housing_df2[['BATHRM', 'BEDRM', 'YEAR', 'PRICE', 'ZIPCODE']]
housing_df3.head(10)
```

```python
housing_df3.dtypes
```

```python
housing_df4 = housing_df3[housing_df3.PRICE.notnull()]
housing_2017 = housing_df4.loc[housing_df4["YEAR"] == 2017.0]
housing_20172 = housing_2017.loc[housing_df4["BEDRM"] == 4]
housing_4bd_2ba = housing_20172.loc[housing_df4["BATHRM"] == 2]
housing_4bd_2ba.head(20)
```

```python
housing_4bd_2ba = housing_4bd_2ba.groupby('ZIPCODE')['PRICE'].mean()
housing_4bd_2ba
```

```python
housing_4bd_2ba.to_excel("4bd2ba.xlsx")
```
